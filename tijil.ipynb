```python
import numpy as np
import pandas as pd
import openpyxl
from pandas import ExcelWriter
from pandas import ExcelFile
import os
```

```python
df=pd.read_csv('Book1.csv')
df = df[0:22]
df
```

```python
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(df[['computation capacity being used']],df['extreme cost'], test_size = 0.3, random_state = 1000)
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 1000, random_state = 1000)
rf.fit(train_features, train_labels)
predictions = rf.predict(test_features)
errors = abs(predictions - test_labels)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'units.')
# rf.score(predictions.reshape(-1,1),test_labels)
yhat = rf.predict(test_features)
SS_Residual = sum((test_labels-yhat)**2)       
SS_Total = sum((test_labels-np.mean(test_labels))**2)     
r_squared = 1 - (float(SS_Residual))/SS_Total
adjusted_r_squared = 1 - (1-r_squared)*(len(test_features)-1)/(len(test_features)-test_features.shape[1]-1)
print(r_squared, adjusted_r_squared)
```
